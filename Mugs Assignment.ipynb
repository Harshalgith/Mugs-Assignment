```python
import json
import csv
from datetime import datetime


def fhir_datetime_to_unix(fhir_datetime):
    """
    Convert FHIR datetime format to Unix timestamp.
    """
    datetime_obj = datetime.strptime(fhir_datetime, '%Y-%m-%dT%H:%M:%S%z')
    return datetime_obj.timestamp()


def create_patient_conditions_map(conditions):
    """
    Create a dictionary with patient_id as key and an array of Condition JSON as value.
    """
    patient_conditions = {}
    for condition in conditions:
        patient_id = condition['subject']['reference'].split('/')[-1]
        if patient_id not in patient_conditions:
            patient_conditions[patient_id] = []
        patient_conditions[patient_id].append(condition)
    return patient_conditions


def create_csv_file(patient_conditions, encounters):
    """
    Create a CSV file with the specified columns.
    """
    with open('output.csv', 'w', newline='') as csvfile:
        fieldnames = ['pid', 'time', 'code', 'description']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for patient_id, conditions in patient_conditions.items():
            for condition in conditions:
                start_time = None
                encounter_id = condition.get('encounter', {}).get('reference', '').split('/')[-1]
                for encounter in encounters:
                    if encounter['id'] == encounter_id:
                        start_time = encounter['period']['start']
                        break
                if start_time:
                    unix_time = fhir_datetime_to_unix(start_time)
                    code = condition['code']['coding'][0]['code']
                    description = condition['code']['coding'][0]['display']
                    writer.writerow({'pid': patient_id, 'time': unix_time, 'code': code, 'description': description})


# Read Patient.ndjson
with open('Patient.ndjson', 'r') as file:
    patient_data = [json.loads(line) for line in file]

# Read Condition.ndjson
with open('Condition.ndjson', 'r') as file:
    condition_data = [json.loads(line) for line in file]

# Read Encounter.ndjson
with open('Encounter.ndjson', 'r') as file:
    encounter_data = [json.loads(line) for line in file]

# Step 1: Extract patient_id as key and array of Condition JSON as value
patient_conditions_map = create_patient_conditions_map(condition_data)

# Step 4: Create a CSV file with the specified columns
create_csv_file(patient_conditions_map, encounter_data)
```

```python
import csv

# Open the CSV file for reading
with open('output.csv', newline='') as csvfile:
    # Create a CSV reader object
    reader = csv.DictReader(csvfile)
    
    # Iterate over each row in the CSV file
    for row in reader:
        # Access data from each row using column names
        patient_id = row['pid']
        timestamp = row['time']
        condition_code = row['code']
        condition_description = row['description']
        
        
        print(f"Patient ID: {patient_id}, Timestamp: {timestamp}, Condition Code: {condition_code}, Description: {condition_description}")
```
